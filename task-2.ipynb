```python
df3 = pd.read_excel('SolarRediation.xlsm')
```

```python
df3.head()
```

```python
df3=df3.drop("UNIXTime", axis=1)
df3=df3.drop("Time", axis=1)
df3=df3.drop("Data", axis=1)
df3=df3.drop("Temperature", axis=1)
df3=df3.drop("WindDirection(Degrees)", axis=1)
df3=df3.drop("Speed", axis=1)
df3=df3.drop("TimeSunRise", axis=1)
df3=df3.drop("TimeSunSet", axis=1)
```

```python
df3['Radiation'] = df3['Radiation'].astype("category").cat.codes
```

```python
df3.head()
```

```python
x=df3.drop('Radiation',axis=1)
y=df3[['Radiation']]
```

```python
x
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=1)
```

```python
x_train
```

```python
x_train_0 = np.c_[np.ones((x_train.shape[0],1)),x_train]
x_test_0 = np.c_[np.ones((x_test.shape[0],1)),x_test]
theta = np.matmul(np.linalg.inv( np.matmul(x_train_0.T,x_train_0) ), np.matmul(x_train_0.T,y_train)) 
```

```python
parameter = ['theta_'+str(i) for i in range(x_train_0.shape[1])]
columns = ['intersect:x_0=1'] + list(x.columns.values)
```

```python
from sklearn.linear_model import LinearRegression
lin_reg = LinearRegression()
lin_reg.fit(x_train,y_train) 
sk_theta = [lin_reg.intercept_]+list(lin_reg.coef_)
```

```python
y_pred_norm =  np.matmul(x_test_0,theta)
#Evaluvation: MSE
J_mse = np.sum((y_pred_norm - y_test)**2)/ x_test_0.shape[0]
# R_square 
sse = np.sum((y_pred_norm - y_test)**2)
sst = np.sum((y_test - y_test.mean())**2)
R_square = 1 - (sse/sst)
print('The Mean Square Error: ',J_mse)
print('R square :',R_square)
```
