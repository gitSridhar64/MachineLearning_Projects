# Section D

```python
staffpershift = pd.read_excel('./Number of staff needed per work shift.xlsx')
```

```python
staffpershift
```

```python
from sklearn import linear_model, metrics
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
```

```python
staffpershift[["Time Period Shifts"]] = staffpershift[["Time Period Shifts"]].apply(lambda col:pd.Categorical(col).codes)
```

```python
x=staffpershift.drop("Number of Staff Required", axis=1)
```

```python
y= staffpershift.drop("Time Period Shifts", axis=1)
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.25,random_state=2)
```

```python
linreg= linear_model.LinearRegression()
```

```python
linreg.fit(x_train, y_train)
```

```python
y_pred=linreg.predict(x_test)
```

```python
y_pred
```

```python
mean_squared_error(y_pred,y_test)
```

```python
linreg.score(y_pred,y_test)
```
