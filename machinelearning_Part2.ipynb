```python
IMG_SIZE= 200
```

```python
data1= train2
```

```python
x_train=[]
y_train=[]
for t in range(len(data1)):
    folder = data1["folder"][t]
    file1 = data1["label"][t]
    image = cv2.imread(os.path.join(folder, file1),cv2.IMREAD_GRAYSCALE)  
    image = cv2.resize(image, (IMG_SIZE, IMG_SIZE))
    image = image/255.0
    image = image.reshape(-1,IMG_SIZE,IMG_SIZE,1)
    image = image.reshape(image.shape[1], image.shape[2])
    if t % 500 == 0:
        print(t)
    x_train.append(image)
    y_train.append(data1["label"][t])
```

```python
x_train = np.array(x_train)
print("x_train done")
y_train = np.array(pd.get_dummies(y_train), dtype= np.uint8)
```

```python
x_train = x_train.reshape(-1,IMG_SIZE,IMG_SIZE, 1)
```

```python
x_test=[]
y_test=[]
for t in range(len(data2)):
    folder = data2["folder"][t]
    file1= data2["label"][t]
    image = cv2.imread(os.path.join(folder, file1),cv2.IMREAD_GRAYSCALE) 
    image = cv2.resize(image, (IMG_SIZE, IMG_SIZE))
    image = image/255.0
    image = image.reshape(1,IMG_SIZE,IMG_SIZE,1) 
    image = image.reshape(image.shape[1], image.shape[2])
    if t % 100 == 0:
        print(t)
    x_test.append(image)
    y_test.append(data2["label"][t])
```

```python
x_test = np.array(x_test)
print("x_test done")
y_test = np.array(pd.get_dummies(y_test), dtype= np.uint8)
```
